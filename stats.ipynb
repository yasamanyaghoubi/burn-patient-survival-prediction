```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("E:\\Uni\\Term 9\\project\\Book1.xlsx")
```

```python
print('Average age is: ',df['Age'].mean())
print('Maximux age is: ',df['Age'].max())
print('Minimum age is: ',df['Age'].min())



# Unique values in the 'Age' column
unique_ages_count = df['Age'].nunique()
print("\nNumber of Unique Ages:", unique_ages_count)

# Visualization: Histogram of ages
plt.figure(figsize=(6, 5))
plt.hist(df['Age'].dropna(), bins=15, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Age Distribution', fontsize=16)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.grid(axis='y', linestyle='--', alpha=0.7)
```

```python
import matplotlib.pyplot as plt

df['Sex'] = df['Sex'].str.strip()

sex_stats = df['Sex'].value_counts()

# Total unique values in the 'Sex' column
unique_sex_count = df['Sex'].nunique()

# Visualization: Bar chart for 'Sex'
plt.figure(figsize=(8, 5))
sex_stats.plot(kind='bar', color=['blue', 'pink'], alpha=0.7, edgecolor='black')
plt.title('Gender Distribution', fontsize=16)
plt.xlabel('Gender', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)


print("Gender Distribution:\n", sex_stats)
print("\nNumber of Unique Genders:", unique_sex_count)
```

```python
# Inspect unique values and their counts in the 'Sex' column

sex_unique_values = df['Sex'].unique()
sex_value_counts = df['Sex'].value_counts()

sex_unique_values, sex_value_counts
```

```python
df['Diagnosis Code'] = df['Diagnosis Code'].str.replace('\n', ', ')  # Replace line breaks with commas
diagnosis_codes = df['Diagnosis Code'].str.split(', ', expand=True).stack()  # Split multiple codes into individual rows

diagnosis_codes_df = diagnosis_codes.reset_index(drop=True).to_frame(name='Code')

diagnosis_code_counts = diagnosis_codes_df['Code'].value_counts()

print("Diagnosis Codes and Their Counts:")
print(diagnosis_code_counts)
```

```python
import matplotlib.pyplot as plt

df['Doctor'] = df['Doctor'].str.strip()

doc_stats = df['Doctor'].value_counts()

unique_doc_count = df['Doctor'].nunique()

plt.figure(figsize=(8, 5))
doc_stats.plot(kind='bar', color=['blue', 'pink'], alpha=0.7, edgecolor='black')
plt.title('Doctor Distribution', fontsize=16)
plt.xlabel('Doctor', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)


# Display statistical data
print("Doctor Distribution:\n", doc_stats)
print("\nNumber of Unique Genders:", unique_doc_count)
```

```python
# Analyze the 'Diagnosis Code' column
diagnosis_code_counts = df['Diagnosis Code'].value_counts()

# Visualization: Bar Chart
plt.figure(figsize=(12, 6))
diagnosis_code_counts.plot(kind='bar', color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Frequency of Diagnosis Codes', fontsize=16)
plt.xlabel('Diagnosis Code', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
```
