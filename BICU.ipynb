```python
from geopy.distance import geodesic

def calculate_distance(city):
    sari_coords = city_coordinates.get("Sari")
    city_coords = city_coordinates.get(city)
    if city_coords:
        return geodesic(sari_coords, city_coords).kilometers
    else:
        return None

```

# Mapping grades to highest

```python
# Function to extract the highest degree from combined values
def simplify_grade(grade):
    if pd.isna(grade):
        return pd.NA
    if isinstance(grade, str):
        grades = [int(g) for g in grade.split(',')]
        return max(grades)  # Keep the highest grade
    return grade  # Return the value directly if not combined
```

```python
# Apply the function to clean the 'Grade' column
df['Grade'] = df['Grade'].apply(simplify_grade)

# Fill missing values in 'Grade' with the mode
grade_mode = df['Grade'].mode()[0]
df['Grade'].fillna(grade_mode, inplace=True)
```
